```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# First let's run every factor, setting koi_disposition as target

#pull target into y; everything else into X
y = df["koi_disposition"]
X = df.drop("koi_disposition", axis=1)
X.head()
```

```python
#since y is text...
y.unique()
```

```python
# let's encode the y target text values
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_y = LabelEncoder().fit(df['koi_disposition'])
y = encoded_y.transform(df['koi_disposition'])
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# split and scale X data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(n_estimators=200)
modelRF.fit(X_train_scaled, y_train)

print(f"Training Data Score: {modelRF.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {modelRF.score(X_test_scaled, y_test)}")
```

```python
# since it's a RandomForrest, let's rank the features
feature_names = X.columns.tolist()
preSelected_features = sorted(zip(modelRF.feature_importances_, feature_names), reverse=True)
ranked_features = pd.DataFrame(preSelected_features, columns=['Score', 'Feature'])
ranked_features = ranked_features.set_index('Feature')
ranked_features
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
#find the 
from pprint import pprint
print('Parameters currently in use:\n')
pprint(modelRF.get_params())
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True],
    'max_depth': [90, 100, 110],
    'max_features': [1, 2, 3],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [4, 6, 8],
    'n_estimators': [100, 200, 300]}

grid = GridSearchCV(estimator = modelRF, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)

print('Random Forests')
print('Test Accuracy: %.3f' % grid.score(X_test, y_test))

# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
